QUESTIONS

```python
#Q1.How does the number of hours studied per week correlate with the final exam score ?
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['Hours_Studied'],data['Exam_Score'],color='red')
plt.title('Hours_Studied vs Final_Exam_Score')
plt.xlabel('Hours_Studied')
plt.ylabel('Final_Exam_Score')
```

```python
#Q2. Is there a significant relationship between class attendance percentage and exam score ?
```

```python
sns.lineplot(data=data,x='Attendance',y='Exam_Score')
plt.title('this is a lineplot')
plt.show()
```

```python
# Q3. does the average number of sleep hours per night influence exam performance?
```

```python
data_groupby = data.groupby('Sleep_Hours')['Exam_Score'].mean()
print(data_groupby)
sns.lineplot(data=data,x='Sleep_Hours',y='Exam_Score')
plt.title('this is a lineplot')
plt.show()
```

```python
#Q4.  How does family income and parental education level impact the student's exam scores?
```
