```python
# Q1. how does the number of hours studied per week corelate with the final exam scores ?
```

```python
plt.figure(figsize=(9,5))
plt.scatter(data['Hours_Studied'],data['Exam_Score'],color='red')
plt.title('Hours_Studied vs Final_exam_score')
plt.xlabel('Hours studied per Week')
plt.ylabel('Final Exam Score')
plt.ylabel('Final Exam Score')
```

```python
# Q2. is there a significant relationship between class attendance percentage and exam scores?
```

```python
data['Attendance'] = pd.to_numeric(data['Attendance'])
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(data[['Attendance','Exam_Score']].corr(),annot=True,cmap='inferno')
plt.title("correlation heatmap")
plt.show()
```
